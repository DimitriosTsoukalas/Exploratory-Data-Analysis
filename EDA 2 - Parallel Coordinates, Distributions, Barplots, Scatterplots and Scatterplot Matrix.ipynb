# EDA 2 - Parallel Coordinates, Distributions, Barplots, Scatterplots and Scatterplot Matrix

## Part 1: Importing database

The iris database will be used in this experiment

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_nparray = iris.data

iris_dataframe = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_dataframe["group"] = pd.Series([iris.target_names[k] for k in iris.target], dtype = "category")
```

## Part 2: Parallel Coordinates

- This graph allows us to identify which groups are easily separable from the other.
- From the below graph we can see that for **setosa**, the petal length is seperable from the other two (versicolor and virginica).

```python
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))

parallel_coordinates(iris_dataframe,"group")
```

## Part 3: Histograms

- Histograms represent the distribution of the whole dataset.
- We can see from the below that **petal width** has two peaks

```python
densityplot = iris_dataframe[iris_dataframe.columns[:4]].plot(kind="density",figsize=(20, 10))
plt.rcParams.update({'font.size': 17})
plt.grid(True)
```

## Part 4: Bar charts

```python
single_dist = iris_dataframe["petal length (cm)"].plot(kind="hist",figsize=(20, 10))
plt.grid(True)
plt.rcParams.update({'font.size': 17})
```

## Part 5: Scatterplot

- When the scatterplots are spread and resembles a line, we can perceive the variables as **correlated**.
- There is a positive correlation between the **petal width** and the **petal length**.
- The division of the cloud indicates that the groups are possuibly seperable.

```python
colors_palete = {'setosa': "red", 'versicolor': "green", 'virginica': "blue"}
colors = [colors_palete[c] for c in iris_dataframe["group"]]
simple_scatterplot = iris_dataframe.plot(kind = "scatter",x="petal length (cm)", y="petal width (cm)", c =colors,figsize=[20,10],s=120)
plt.grid(True)
```

## Part 6 Scatterplot Matrix

- **ONLY** because the number of variables is **NOT too large** , we can generate a **matrix of scatterplots**.

```python
from pandas.plotting import scatter_matrix

colors_palete = {'setosa': "red", 'versicolor': "green", 'virginica': "blue"}
colors = [colors_palete[c] for c in iris_dataframe["group"]]
martix_of_scatterplots = scatter_matrix(iris_dataframe,figsize=(15,15),color=colors,diagonal="kde")
plt.rcParams.update({'font.size': 10})
```
